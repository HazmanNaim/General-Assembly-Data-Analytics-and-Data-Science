# Linear Regression for Diabetes Progression

1. Import modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes #sample dataset
from sklearn.model_selection import train_test_split #split training and testing data
from sklearn.linear_model import LinearRegression #machine learning
from sklearn.metrics import mean_squared_error #model evaluation
```

2. Load diabetes dataset (load_diabetes) from sklearn.datasets, define a dataframe and split input (X) and target (y)

```python
diabetes = load_diabetes()

# Convert to DataFrame
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)

df.head()
```

The diabetes dataset from scikit-learn includes ten baseline variables, age, sex, BMI, average blood pressure, and six blood serum measurements, which are:
1. Age: Age in years.
2. Sex: Gender (1 = male, 0 = female).
3. BMI: Body mass index, a measure of body fat based on height and weight.
4. Average Blood Pressure: Mean arterial blood pressure.
5. S1 to S6: Six blood serum measurements.      
- S1: T-Cells (a type of white blood cell) in the blood.   
- S2: Low-density lipoproteins (a type of cholesterol) in the blood.   
- S3: High-density lipoproteins (a type of cholesterol) in the blood.   
- S4: Total cholesterol in the blood.   
- S5: Triglycerides in the blood.   
- S6: Blood sugar level.
6. Target: A quantitative measure of disease progression one year after baseline.

```python
# Get the target values
target = diabetes.target

# Now 'target' contains the target values
target
```

3. Split the dataset into training and testing sets (20% test and 80% training)

```python
X = diabetes.data
Y = diabetes.target

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

4. Create and train the linear regression model

```python
# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

5. Make predictions using the trained model

```python
# Make predictions using the trained model
y_pred = model.predict(X_test)
```

6. Plot predicted vs. actual values

```python
# Plot predicted vs. actual values
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.title("Actual VS Predicted")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

7. Calculate three evaluation measures

```python
mse = mean_squared_error(y_test, y_pred)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("-"*30)
print("Evaluation Metrics")
print("-"*30)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Root Mean Squared Error:", np.sqrt(mse))
print("R-squared:", r2_score(y_test, y_pred))
```
